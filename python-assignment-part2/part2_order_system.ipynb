```python
#Extracting distinct categories
categories = []

for item in menu:
    cat = menu[item]["category"]
    if cat not in categories:
        categories.append(cat)

#Printing grouped menu
for cat in categories:
    print(f"\n===== {cat} =====")
    
    for item in menu:
        details = menu[item]
        
        if details["category"] == cat:
            price = details["price"]
            status = "Available" if details["available"] else "Unavailable"
            
            print(f"{item:<18} ₹{price:.2f}   [{status}]")
```

```python
# Menu Metrics

# Total items
total_items = len(menu)

# Total available items
available_count = 0

for item in menu:
    if menu[item]["available"]:
        available_count += 1

# Most expensive item
max_price = 0
max_item = ""

for item in menu:
    price = menu[item]["price"]
    
    if price > max_price:
        max_price = price
        max_item = item

# Items under ₹150
cheaper_items = []

for item in menu:
    price = menu[item]["price"]
    
    if price < 150:
        cheaper_items.append((item, price))
```

```python
# Printing the above outputs

print("\n--- Menu Summary ---")

print("Total items on menu       :", total_items)
print("Total available items     :", available_count)

print("\nMost expensive item:")
print(max_item, "₹", max_price)

print("\nItems under ₹150:")
for item, price in cheaper_items:
    print(f"{item:<16} ₹{price:.2f}")
```

# **#Task 2 — Cart Operations**

```python
cart = []
```

```python
# Adding item logic definition

def add_to_cart(item_name, qty):
    
    # Check if item exists
    if item_name not in menu:
        print(f"⚠ {item_name} not found in menu")
        return
    
    # Check availability
    if not menu[item_name]["available"]:
        print(f"⚠ {item_name} is currently unavailable")
        return
    
    # Check if already in cart
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += qty
            print(f"✓ Updated {item_name} quantity to {entry['quantity']}")
            return
    
    # Add new entry
    cart.append({
        "item": item_name,
        "quantity": qty,
        "price": menu[item_name]["price"]
    })
    
    print(f"✓ Added {item_name} x{qty}")
```

```python
# Removing item logic definition

def remove_from_cart(item_name):
    
    found = False
    
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            found = True
            print(f"✓ Removed {item_name}")
            break
    
    if not found:
        print(f"{item_name} not in cart")
```

```python
# Updating quantity logic definition

def update_quantity(item_name, new_qty):
    
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = new_qty
            print(f"✓ Updated {item_name} to {new_qty}")
            return
    
    print(f"{item_name} not found in cart")
```

```python
# Print cart logic definition

def print_cart():
    print("\nCurrent Cart:")
    
    if not cart:
        print("(empty)")
        return
    
    for entry in cart:
        print(entry)
```

```python
# Test cases

# 1
add_to_cart("Paneer Tikka", 2)
print_cart()

# 2
add_to_cart("Gulab Jamun", 1)
print_cart()

# 3
add_to_cart("Paneer Tikka", 1)
print_cart()

# 4
add_to_cart("Mystery Burger", 1)
print_cart()

# 5
add_to_cart("Chicken Wings", 1)
print_cart()

# 6
remove_from_cart("Gulab Jamun")
print_cart()
```

```python
# Printing order summary

print("\n========== Order Summary ==========")

subtotal = 0

for entry in cart:
    item = entry["item"]
    qty = entry["quantity"]
    price = entry["price"]
    
    total = qty * price
    subtotal += total
    
    print(f"{item:<18} x{qty}    ₹{total:.2f}")

print("------------------------------------")

gst = round(subtotal * 0.05, 2)
total_payable = round(subtotal + gst, 2)

print(f"Subtotal:                ₹{subtotal:.2f}")
print(f"GST (5%):               ₹{gst:.2f}")
print(f"Total Payable:          ₹{total_payable:.2f}")
print("====================================")
```

# **#Task 3 — Inventory Tracker with Deep Copy**

```python
import copy

# Create deep copy
inventory_backup = copy.deepcopy(inventory)

# Modify original inventory (demo)
inventory["Paneer Tikka"]["stock"] -= 2

print("After manual change:\n")

print("Inventory (modified):")
print(inventory["Paneer Tikka"])

print("\nInventory Backup (unchanged):")
print(inventory_backup["Paneer Tikka"])
```

```python
# Restore inventory to original state
inventory = copy.deepcopy(inventory_backup)

print("\nInventory restored:")
print(inventory["Paneer Tikka"])
```

```python
#Deduct stock based on cart items

print("\n--- Order Fulfilment ---")

for entry in cart:
    item = entry["item"]
    qty_needed = entry["quantity"]
    
    current_stock = inventory[item]["stock"]
    
    if current_stock >= qty_needed:
        inventory[item]["stock"] -= qty_needed
        print(f"✓ {item}: Deducted {qty_needed}, Remaining {inventory[item]['stock']}")
    
    else:
        print(f"⚠ {item}: Only {current_stock} available, cannot fulfill full order")
        inventory[item]["stock"] = 0
```

```python
# Check for reorder alerts

print("\n--- Reorder Alerts ---")

for item in inventory:
    stock = inventory[item]["stock"]
    reorder_level = inventory[item]["reorder_level"]
    
    if stock <= reorder_level:
        print(f"⚠ Reorder Alert: {item} — Only {stock} unit(s) left (reorder level: {reorder_level})")
```

```python
# Final comparison of inventory states

print("\n--- Final Comparison ---")

print("\nCurrent Inventory (after order):")
for item in inventory:
    print(item, ":", inventory[item])

print("\nBackup Inventory (original):")
for item in inventory_backup:
    print(item, ":", inventory_backup[item])
```

# **#Task 4— Daily Sales Log Analysis**

```python
# Printing revenue per day from sales log

print("Revenue Per Day:\n")

daily_revenue = {}

for date in sales_log:
    total = 0
    
    for order in sales_log[date]:
        total += order["total"]
    
    daily_revenue[date] = total
    print(f"{date} : ₹{total:.2f}")
```

```python
# Identify best-selling day   

max_revenue = 0
best_day = ""

for date in daily_revenue:
    if daily_revenue[date] > max_revenue:
        max_revenue = daily_revenue[date]
        best_day = date

print("\nBest-Selling Day:")
print(best_day, "with ₹", round(max_revenue, 2))
```

```python
# Identify most ordered item

item_count = {}

for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            
            if item not in item_count:
                item_count[item] = 1
            else:
                item_count[item] += 1

# Find max
max_count = 0
top_item = ""

for item in item_count:
    if item_count[item] > max_count:
        max_count = item_count[item]
        top_item = item

print("\nMost Ordered Item:")
print(top_item, "ordered", max_count, "times")
```

```python
# Adding a new day to sales log

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("\nNew day added: 2025-01-05")
```

```python
# Recompute revenue per day with new data

print("\nUpdated Revenue Per Day:\n")

daily_revenue = {}

for date in sales_log:
    total = 0
    
    for order in sales_log[date]:
        total += order["total"]
    
    daily_revenue[date] = total
    print(f"{date} : ₹{total:.2f}")

# Recompute best day
max_revenue = 0
best_day = ""

for date in daily_revenue:
    if daily_revenue[date] > max_revenue:
        max_revenue = daily_revenue[date]
        best_day = date

print("\nUpdated Best-Selling Day:")
print(best_day, "with ₹", round(max_revenue, 2))
```

```python
#Enumerated order list

print("\nAll Orders:\n")

order_list = []

# Flatten orders
for date in sales_log:
    for order in sales_log[date]:
        order_list.append((date, order))

# Enumerate
for idx, (date, order) in enumerate(order_list, start=1):
    
    items_str = ", ".join(order["items"])
    
    print(f"{idx}.  [{date}] Order #{order['order_id']}  — ₹{order['total']:.2f} — Items: {items_str}")
```
